```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##### Data is downloaded from MovieLens https://grouplens.org/datasets/movielens/, 'MovieLens 1M Dataset' tab
uname=('UserID','Gender','Age','Occupation','Zipcode')
dt1=pd.read_table('D:/pyLX/ml-1m/users.dat',sep='::',header=None,names=uname)

rname=('UserID','Movie_ID','Rating','Timestamp')
dt2=pd.read_table('D:/pyLX/ml-1m/ratings.dat',sep='::',header=None,names=rname)

mname=('Movie_ID','Title','Genres')
dt3=pd.read_table('D:/pyLX/ml-1m/movies.dat',sep='::',header=None,names=mname)
```

```python
dt=pd.merge(pd.merge(dt1,dt2),dt3)
print(len(dt))
dt.head()
```

```python
##### Movies that most people have seen
dt.groupby(dt['Title']).size().sort_values(ascending=False)
```

```python
##### Movies that are received high review scores
#dt.pivot_table(values='Rating',index='Title',aggfunc='mean').sort_values(by='Rating',ascending=False)
dt['Rating'].groupby(dt['Title']).mean().sort_values(ascending=False)
```

```python
##### Movies that most people have seen have high review scores?
seen=dt.groupby(dt['Title']).size()
popular=dt.groupby(dt['Title']).size().sort_values(ascending=False).head(10)
rating=dt['Rating'].groupby(dt['Title']).mean()
top=dt['Rating'].groupby(dt['Title']).mean().sort_values(ascending=False).head(16)
```

```python
rating[popular.index]
```

```python
##### Movies with high review scores are the popular ones?
seen[top.index]
```

```python
##### Define: movies with high review scores & being popular are the best
hot=seen[seen>1000]
rating[hot.index]
```

```python
##### Is there a gender bias towards movie reviews?
df=dt[dt['Title'].isin(hot.index)]
df.pivot_table(index='Title',columns='Gender',aggfunc='count')
```

```python
rt=df.pivot_table(values='Rating',index='Title',columns='Gender',aggfunc='mean')
rt['Dif_R']=rt.F-rt.M
rt.sort_values(by='Dif_R',ascending=True).head(10)
rt.sort_values(by='Dif_R',ascending=False).head(10)
```
